### Realtime video demo for Face Emotion Recognition

```python
import cv2
from deepface import DeepFace

faceCascade=cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap= cv2.VideoCapture(-1)
if not cap.isOpened():
    cap=cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

happy=0
total =0
while True:
    ret,frame=cap.read()
    result=DeepFace.analyze(frame, actions=["emotion"],enforce_detection=False)
    
    gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces=faceCascade.detectMultiScale(gray,1.1,4)
    
    for(x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w, y+h),(0,255,0),2)
        
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,
            result["dominant_emotion"],
            (50,100),
            font, 3,
            (0,0,255),
            2,
            cv2.LINE_4
           )
    if(result["dominant_emotion"]=="happy"):
        happy=happy+1;
    total=total+1
    if not ret:
        break
    cv2.imshow('asd', frame)
    cv2.waitKey(1)
    if cv2.getWindowProperty('asd', 4) < 1:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from deepface import DeepFace

faceCascade=cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap= cv2.VideoCapture(-1)
if not cap.isOpened():
    cap=cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

happy=0
total =0
while True:
    ret,frame=cap.read()
    result=DeepFace.analyze(frame, actions=["emotion"],enforce_detection=False)
    
    gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces=faceCascade.detectMultiScale(gray,1.1,4)
    
    for(x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w, y+h),(0,255,0),2)
        
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,
            result["dominant_emotion"],
            (50,100),
            font, 3,
            (0,0,255),
            2,
            cv2.LINE_4
           )
    if(result["dominant_emotion"]=="angry"):
        angry=angry+1;
    elif(result["dominant_emotion"]=="disgust"):
        disgust=disgust+1;
    elif(result["dominant_emotion"]=="fear"):
        fear=fear+1;
    elif(result["dominant_emotion"]=="happy"):
        happy=happy+1;
    elif(result["dominant_emotion"]=="sad"):
        sad=sad+1;
    elif(result["dominant_emotion"]=="surprise"):
        surprise=surprise+1;
    elif(result["dominant_emotion"]=="neutral"):
        neutral=neutral+1;
    total=total+1
    if not ret:
        break
    cv2.imshow('asd', frame)
    cv2.waitKey(1)
    if cv2.getWindowProperty('asd', 4) < 1:
        break
cap.release()
cv2.destroyAllWindows()

ratios=[angry/total,disgust/total,fear/total,happy/total,sad/total,surprise/total,neutral/total]

with open('comedy.csv', 'a', newline='') as f_object:  
    # Pass the CSV  file object to the writer() function
    writer_object = writer(f_object)
    # Result - a writer object
    # Pass the data in the list as an argument into the writerow() function
    writer_object.writerow(ratios)  
    # Close the file object
    f_object.close()
    
angry=0
disgust=0
fear=0
happy=0
sad=0
surprise=0
neutral=0
total=0
```
