```python
from csv import writer
import cv2
from deepface import DeepFace

faceCascade=cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap= cv2.VideoCapture(-1)
if not cap.isOpened():
    cap=cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

angry=0
disgust=0
fear=0
happy=0
sad=0
surprise=0
neutral=0
total=0

while True:
    ret,frame=cap.read()
    result=DeepFace.analyze(frame, actions=["emotion"],enforce_detection=False)
    
    gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces=faceCascade.detectMultiScale(gray,1.1,4)
    
    for(x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w, y+h),(0,255,0),2)
        
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,
            result["dominant_emotion"],
            (50,100),
            font, 3,
            (0,0,255),
            2,
            cv2.LINE_4
           )
    if(result["dominant_emotion"]=="angry"):
        angry=angry+1;
    elif(result["dominant_emotion"]=="disgust"):
        disgust=disgust+1;
    elif(result["dominant_emotion"]=="fear"):
        fear=fear+1;
    elif(result["dominant_emotion"]=="happy"):
        happy=happy+1;
    elif(result["dominant_emotion"]=="sad"):
        sad=sad+1;
    elif(result["dominant_emotion"]=="surprise"):
        surprise=surprise+1;
    elif(result["dominant_emotion"]=="neutral"):
        neutral=neutral+1;
    total=total+1
    if not ret:
        break
    cv2.imshow('asd', frame)
    cv2.waitKey(1)
    if cv2.getWindowProperty('asd', 4) < 1:
        break
cap.release()
cv2.destroyAllWindows()
enj=int(input("If you enjoyed the video, input 1, else type 0 :"))
other_str = (str(input("If you would like to add in other types \(Such as \n action,comedy,drama,horror,romance \) then \n input them seperated with a \",\" : ")))
ratios = [angry/total,disgust/total,fear/total,happy/total,sad/total,surprise/total,neutral/total,total]
other_arr = other_str.split(',')
types_arr=['action','comedy','drama','horror','romance']
types_num_arr=[0,0,0,1,0]
for i in range(len(types_arr)):
    for j in range(len(other_arr)):
        if(types_arr[i]==other_arr[j]):
            types_num_arr[i]=1
            break
            
ratios_plus_types=[angry/total,disgust/total,fear/total,happy/total,sad/total,surprise/total,neutral/total]
for i in range(len(types_num_arr)):
    ratios_plus_types.append(types_num_arr[i])
                   
ratios_plus_types.append(total)
ratios_plus_types.append(enj)
ratios.append(enj)   
                   
with open('horror.csv', 'a', newline='') as f_object:  
    # Pass the CSV  file object to the writer() function
    writer_object = writer(f_object)
    # Result - a writer object
    # Pass the data in the list as an argument into the writerow() function
    writer_object.writerow(ratios)  
    # Close the file object
    f_object.close()


with open('all.csv', 'a', newline='') as f_object:  
    # Pass the CSV  file object to the writer() function
    writer_object = writer(f_object)
    # Result - a writer object
    # Pass the data in the list as an argument into the writerow() function
    writer_object.writerow(ratios_plus_types)  
    # Close the file object
    f_object.close()
    
```
